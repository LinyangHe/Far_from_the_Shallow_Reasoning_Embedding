```python
import json
import os
import re
```

```python
directory = "/home/lh3288/project/PJ2504_Dual_stream/LLM_Dual_Stream/LLM_Probing/data/Logic-LLM/"
json_files = [f for f in os.listdir(directory) if f.endswith(".json")]
json_files = ['ProntoQA_dev.json']
json_files
```

```python
merged_qa = []

for file_name in json_files:
    with open(directory + file_name, "r") as f:
        data = json.load(f)
        for qa in data:
            if qa["answer"] not in {"A", "B"}:
                continue
            else:
                del qa["options"]

                question = qa["question"]
                qa["question"] = re.sub(r'true,\s*false,\s*or\s*[^?]*\?', 'true or false?', question)

                text_answer = qa["answer"]
                qa["answer"] = True if text_answer == "A" else False
                merged_qa.append(qa)

# with open(directory + "logic_llm.jsonl", "w") as f:
#     for qa in merged_qa:
#         f.write(json.dumps(qa) + "\n")
```

```python
yes_true = yes_false = not_true = not_false = total = 0

for item in merged_qa:
    question = item["question"]
    answer = item["answer"]
    
    total += 1
    if "not" in question and answer:
        not_true += 1
    elif "not" in question and not answer:
        not_false += 1
    elif not "not" in question and answer:
        yes_true += 1
    elif not "not" in question and not answer:
        yes_false += 1

print(total, yes_true, yes_false, not_true, not_false)
```
